```python
from sympy import *
from functools import reduce
import numpy as np
from scipy.optimize import fmin
```

```python
n = int(input("Digite quantas variáveis sua função possui: "))
symx = symbols(f"x(1:{n+1})")

f = str(input("Digite a sua função: "))
fx = simplify(f)

a = symbols("a", real = True)

x0 = list(map(int,input("Digite o seu ponto x0 (ex: 1 2): ").strip().split()))[:n]
```

```python
def gradiente(fx, symx):
    g = []
    for x in symx:
        g.append(diff(fx, x))
    return(g)
```

```python
#Calculo do gradiente da F: 
sg = gradiente(fx, symx)
sg
```

```python
#inicializar H com a matriz identidade
hk = eye(n)
hk
```

```python
def subsGrad(gx, symx, x0):
    subsFx = gx[:]
    for i in range(len(subsFx)):
        for j in range(len(symx)):
            subsFx[i] = subsFx[i].subs(symx[j], x0[j])
        
    return subsFx
```

```python
def subsFx(fx, symx, x):
    subsFx = fx
    for i in range(len(symx)):
        subsFx = subsFx.subs(symx[i], x[i])

    return subsFx
```

```python
#substituir o ponto xk no gradiente de F:
gk = subsGrad(sg, symx, x0)
gk 
```

```python
def critParada(gk):
    elevQuadrado = list(map(lambda x: x ** 2, gk))
    soma = sum(elevQuadrado)
    raiz = sqrt(soma)
    return raiz    
```

```python
def calculaCDFP(vk, rk, hk):
    part1 = np.zeros((2,2))
    part2 = np.zeros((2,2))
    part3 = np.zeros((2,2))
    part4 = np.zeros((2,2))
    part5 = np.zeros((2,2))
    part6 = np.zeros((2,1))
    part7 = np.zeros((1,1))
    part8 = np.zeros((2,2))
    part9 = np.zeros((2,2))

    part1 = np.outer(vk,vk.T)
    part2 = np.outer(vk.T,rk)
    part3 = np.dot(part1, 1/part2)

    part4 = np.outer(rk,rk.T)
    part5 = np.dot(hk,part4)
    part6 = np.dot(part5,hk)
    part7 = np.dot(rk.T,hk)
    part8 = np.dot(part7,rk)
    part9 = np.dot(part6,1/part8)

    CDFP = part3 - part9
    return CDFP
```

```python
def calculaCBFGS(vk, rk, hk):
    part1 = np.zeros((2,2))
    part2 = np.zeros((2,2))
    part3 = np.zeros((2,2))
    part4 = np.zeros((2,2))
    part5 = np.zeros((2,2))
    part6 = np.zeros((2,2))
    part7 = np.zeros((2,2))
    part8 = np.zeros((2,2))
    part9 = np.zeros((2,2))
    part10 = np.zeros((2,2))
    part11 = np.zeros((2,2))
    part12 = np.zeros((2,2))
    part13 = np.zeros((2,2))
    part14 = np.zeros((2,2))
    part15 = np.zeros((2,2))
    part16 = np.zeros((2))

    part1 = np.outer(rk.T, rk)
    part2 = np.dot(part1, hk)
    part3 = np.outer(rk.T, vk)
    part4 = np.dot(part2, 1/part3)
    part5 = 1 + part4

    part6 = np.outer(vk, vk.T)
    part7 = np.outer(vk.T, rk)
    part8 = np.dot(part6, 1/part7)

    part9 = np.dot(part5, part8)

    part10 = np.outer(vk, rk.T)
    part11 = np.dot(part10, hk)

    part12 = np.outer(rk, vk.T)
    part13 = np.dot(part12, hk)

    part14 = part13 + part11

    part15 = np.outer(rk.T, vk)

    part16 = np.dot(part14, 1/part15)

    CBFGS = part9 - part16
    return(CBFGS)
```

```python
k = 0
x = [x0, ]
vk = np.zeros((2,1))
rk = np.zeros((2,1))

while  critParada(gk) > 0.001:
#while k < 1:
    X = x[k] - a * np.dot(gk, hk)
    #print("-"*50)
    #print(f"Fx: {fx}")
    #print(f"X: {X}")

    FAlfa = subsFx(fx, symx, X)

    #print(f"fAlfa: {FAlfa}")

    difFAlfa = diff(FAlfa, a)

    print("-" * 50)
    print(f"d/da: {difFAlfa}")

    solveFAlfa = nroots(difFAlfa, n=15)
    
    print(f"Alfas: {solveFAlfa}")
    
    valorAlfa = solveFAlfa[0]
    print(f"Alfa: {valorAlfa}")

    for i in solveFAlfa:
        if FAlfa.subs(a, i) < FAlfa.subs(a, valorAlfa):
            valorAlfa = i
    
    #print(f"fAlfa: {FAlfa}")
    #valorAlfa = fmin(lambdify(a, FAlfa), 1.0, disp = 0)

    
    print(f"Alfa: {valorAlfa}")
    print(f"Xbefore: {X}")
    
        
    for i in range(len(X)):
        X[i] = X[i].subs(a, valorAlfa)

    G = subsGrad(sg, symx, X)

    
    print(f"X: {X}")
    print(f"Xk: {x[k]}")
    print(f"G: {G}")
    print(f"Gk: {gk}")

    
    
    vk = list(map(lambda i, j: i - j, x[k], X))
    #vk = x[k] - X
    
    rk = list(map(lambda i, j: i - j, gk, G))
    #rk = gk - G

    vk = np.array(vk)
    rk = np.array(rk)
    hk = np.array(hk)

    print(f"vk: {vk}")
    print(f"rk: {rk}")
    print(f"hk: {hk}")

    CDFP = calculaCDFP(vk, rk, hk)
    CBFGS = calculaCBFGS(vk, rk, hk)
        
    print(f"CDFP: {CDFP}")
    print(f"CBFGS: {CBFGS}")

    beta = np.random.rand()

    ck = (1 - beta) * CDFP + beta * CBFGS

    H = list(map(lambda i, j: i + j, hk, ck))

    k += 1

    x.append(X)
    gk = G
    hk = H


print(F"Ponto ótimo encontrado: {x[-1]}")
```
